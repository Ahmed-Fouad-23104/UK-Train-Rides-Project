```
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```
ticket_df = pd.read_excel("/content/Railway Final Dataset (1) (1).xlsx", sheet_name="TicketInfo")
```

```
ticket_df["Date_of_Purchase"] = pd.to_datetime(ticket_df["Date_of_Purchase"])
ticket_df["Month"] = ticket_df["Date_of_Purchase"].dt.month
ticket_df["Year"] = ticket_df["Date_of_Purchase"].dt.year
```

```
journey_monthly = (
    ticket_df.groupby(["Year", "Month"])
    .size()
    .reset_index(name="Journey_Count")
)

journey_monthly
```

```
train_df = journey_monthly[journey_monthly["Month"] <= 4]

X = train_df[["Month"]]                 # Feature
y = train_df["Journey_Count"]           # Target
```

```
model_journey = RandomForestRegressor()
model_journey.fit(X, y)

# Predict on training data
pred_train = model_journey.predict(X)

acc_journey = r2_score(y, pred_train) * 100

print("===== MODEL ACCURACY =====")
print(f"Journey Model Accuracy: {acc_journey:.2f}%")
```

```
may = pd.DataFrame({"Month": [5]})

predicted_journey = model_journey.predict(may)[0]

print("===== PREDICTED JOURNEYS FOR MAY =====")
print(f"Predicted Journey Count: {predicted_journey:.0f} journeys")
```
